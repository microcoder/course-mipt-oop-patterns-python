# Профессионализм в программировании

## Профессионал

Профессиональный программист отличается от любителя тем, что он:

* Адекватно оценивает
  * Срок создания программного продукта
  * Необходимые для этого ресурсы
* Хорошо планирует
  * Архитектуру программного продукта
  * Последовательность разработки
* Выполняет работу качественно
  * ПО надёжно функционирует
  * Программный код легко читать и сопровождать
  
## Каскадная модель разработки ПО (Waterfall - модель водопада)

* Анализ требований к ПО
* Разработка архитектуры программы
* Кодирование
* Тестирование и отладка
* Инсталляция и поддержка

## Построение архетектуры ПО

Проектирование архитектуры программы помогает понять, что именно надо делать и когда. Профессиональному программисту важно знать:

* Парадигмы программирования
  * Структурная парадигма (движение «сверху-вниз»)
  * Модульная парадигма (разбиение программы на модули)
  * Объектно-ориентированное проектирование
* Готовые архитектурные решения - паттерны проектирования
* Визуализация архитектуры ПО

## Ясность и чистота исходных текстов программы

* Читабельность кода (readability) даёт возможность:
  * Быстро понимать смысл исходного текста
  * Лучше видеть ошибки
  * Увереннее модифицировать код
  * Уменьшить объём внутренней документации

## Качество работы программы

* Делает то, что заявлено в техническом задании
* Делает это без ошибок
* Выполняет свои функции быстро
* Нравится конечному пользователю
