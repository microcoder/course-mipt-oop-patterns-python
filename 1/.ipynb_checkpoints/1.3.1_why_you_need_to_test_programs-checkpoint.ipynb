# Зачем нужно тестировать программы

## Обратная связь

* Положительная обратная связь:
    * Положительные отзывы конечных пользователей
    * Запросы пользоателей на новую функциональность
    * Увеличение объёма продаж
* Отрицательная обратная связь:
    * Негативные отзывы конечных пользователей
    * Отсутствие интереса к программному продукту
    * Падение объёма продаж

## Тестирование

Тестировщик даёт участникам проекта по разработке ПО отрицательную обратную связь о качестве программного продукта на самой ранней стадии, когда ещё не поздно всё исправить.

* **Тестирование** (Quality Control) - это:
    * проверка соответствия между реальным поведением программы и её ожидаемым поведением
    * на конечном наборе тестов, выбранном определенным образом
    
    Источник: IEEE Guide to Software Engineering Body of Knowledge, SWEBOK, 2014
    
## Обязанности тестировщика

* Находить дефекты ("баги")
* Вносить описание найденного дефекта в систему отслеживания ошибок (Bug tracking system)
* Описывать способы воспроизведения ошибок
* Создавать отчеты о тестировании

А также, не обязан, но может:

* Исправлять документацию
* Анализировать и уточнять требования к программе
* Создавать ПО для автоматизации процесса тестирования
* Результат работы тестировщика - повышение качества прогрммного продукта

## Аспекты качества программы

* Функциональность
    * Пригодность к использованию
    * Правильность выполнения задачи
    * Поддержка стандартов
    * Защищённость (security)
* Надёжность
    * Низкая частота отказов
    * Отказоустойчивость
    * Способность к восстановлению
* Практичность
    * Понятность в использовании
    * Управляемость
    * Привлекательность
* Эффективность
    * Время отклика программы
    * Объём использования ресурсов ПК
* Сопровождаемость (поддержка)
* Переносимость (кроссплатформенность)

## Классификация тестирования по масштабу

* Модульное тестирование (unittest)
    * Тестирование отдельных операций, методов, функций
* Интеграционное тестирование
    * Проверка, что модули взаимодействуют между собою корректно
* Системное тестирование
    * Тестирование на уровне пользовательского интерфейса

## Совместимость ролей

* Может ли быть тестировщиком:
    * Программист - <span style="color: red;">нет</span>
    * Старший программист (Team leader) - <span style="color: red;">нет</span>
    * Менеджер проекта - <span style="color: green;">да</span>
    * Технический писатель - <span style="color: green;">да</span>
    * Эксперт предметной области - <span style="color: green;">да</span>
    * Представитель заказчика - <span style="color: green;">да</span>
* Методика [TDD](https://ru.wikipedia.org/wiki/%D0%A0%D0%B0%D0%B7%D1%80%D0%B0%D0%B1%D0%BE%D1%82%D0%BA%D0%B0_%D1%87%D0%B5%D1%80%D0%B5%D0%B7_%D1%82%D0%B5%D1%81%D1%82%D0%B8%D1%80%D0%BE%D0%B2%D0%B0%D0%BD%D0%B8%D0%B5) позволяет программисту быть самому себе эффективным тестировщиком

## Отладка программы

1. Уяснить суть ошибки и обнаружить её причину
2. Локализовать её в исходном тексте программы
3. Устранить ошибку


* Типичные задачи отладки:
    * Узнавать текущие значения переменных
    * Выяснять, по какому пути выполнялась программа

* Два пути отладки:
    * Использование отладчиков (debugger)
    * Логгирование (вывод отладочных сведений)
